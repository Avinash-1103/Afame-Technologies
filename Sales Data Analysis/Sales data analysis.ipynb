```python
# Grouping data by region
region_sales_profit = data.groupby('Region').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
```

```python
# Setting the positions and width for the bars
bar_width = 0.35
index = np.arange(len(region_sales_profit['Region']))

# Creating the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting sales bars
bar1 = ax.bar(index, region_sales_profit['Sales'], bar_width, label='Sales')

# Plotting profit bars
bar2 = ax.bar(index + bar_width, region_sales_profit['Profit'], bar_width, label='Profit')

# Adding title and labels
ax.set_title('Total Sales and Profit by Region')
ax.set_xlabel('Region')
ax.set_ylabel('Amount')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(region_sales_profit['Region'])

# Adding the legend
ax.legend()

# Displaying the plot
plt.tight_layout()
plt.show()
```

Analyzing Most selling product of the Year

```python
product_sales_by_year = pd.pivot_table(
    data,
    values='Sales',
    index='Product Name',
    columns='Year',
    aggfunc='sum',
    fill_value=0
)
```

```python
product_sales_by_year
```

```python
top_products_years = data.groupby(['Year','Product Name'])['Sales'].sum().reset_index()
```

```python
top_products_years.head()
```

```python
top_products_by_year = top_products_years.groupby('Year').apply(
    lambda x: x.nlargest(5, 'Sales')
).reset_index(drop=True)

# Display the top products by year
top_products_by_year.head(15)
```

```python
years = top_products_by_year['Year'].unique()
```

```python
# Set up the figure and axes
fig, axs = plt.subplots(len(years), 1, figsize=(10, 5 * len(years)))

# Plot data on each subplot
for i, year in enumerate(years):
    # Filter data for the specific year
    yearly_data = top_products_by_year[top_products_by_year['Year'] == year]

    # Plot each top product's sales
    for product in yearly_data['Product Name']:
        product_data = data[(data['Year'] == year) & (data['Product Name'] == product)]
        monthly_sales = product_data.groupby(product_data['Order Date'].dt.month)['Sales'].sum()

        axs[i].plot(monthly_sales.index, monthly_sales.values, marker='o', label=product)

    axs[i].set_title(f'Top Products of {year}')
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('Sales')
    axs[i].legend()
    axs[i].set_xticks(range(1, 13))

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
product_sales_by_month = pd.pivot_table(
    data,
    values='Sales',
    index=['Year', 'Month'],
    columns='Product Name',
    aggfunc='sum',
    fill_value=0
)

# Compute total sales for each product
total_sales_by_product = data.groupby('Product Name')['Sales'].sum()

# Identify the top 5 products by total sales
top_5_products = total_sales_by_product.nlargest(5).index

# Plotting
fig, axs = plt.subplots(5, 1, figsize=(10, 20))

for i, product in enumerate(top_5_products):
    ax = axs[i]
    product_sales = product_sales_by_month[product]
    product_sales = product_sales.reset_index()
    ax.plot(product_sales.index, product_sales[product], marker='o', label=product)
    ax.set_title(f'Sales Trend for {product}')
    ax.set_xlabel('Month-Year')
    ax.set_ylabel('Sales')
    ax.legend()
    # ax.tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Clustering customers based on their purchasing behavior
from sklearn.cluster import KMeans
customer_data = data.groupby('Customer ID').agg({'Sales': 'sum', 'Order ID': 'count'}).rename(columns={'Order ID': 'Order Count'})
kmeans = KMeans(n_clusters=3)
customer_data['Cluster'] = kmeans.fit_predict(customer_data)
print(customer_data.head())
```
